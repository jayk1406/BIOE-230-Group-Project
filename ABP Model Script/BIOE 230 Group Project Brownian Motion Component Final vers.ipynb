```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.random import random as rand
from scipy import stats
```

```python
def get_trajectory(num_steps, Dtheta, vel):
    #time step, translation diffusion (optional)
    dt = 0.01; Dtrans = 0.0;
    xvec=np.zeros(0); yvec=np.zeros(0); thetavec = np.zeros(0)
    x=0.0; y = 0.0; theta = 0.0
    for i in range(num_steps):
        # diffusive/random steps
        dx = np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
        dy= np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
        dtheta = np.sqrt(2*Dtheta*dt)*2*(rand(1)-0.5);
        # update coordinates (including ballistic step)
        x += vel*dt*np.cos(theta) + dx 
        y += vel*dt*np.sin(theta) + dy
        xvec = np.append(xvec,x); yvec = np.append(yvec,y)
        theta += dtheta
        thetavec = np.append(thetavec, theta)
    return xvec, yvec, thetavec
```

```python
#ABP model parameters: ballistic velocity, time step, rotational diffusion constant, translational diffusion constant
vel =2.0; dt = 0.01; Drot = 1.0 ; Dtrans = 0.1; 
#boundary size parameter
Bbox = 20
Bframe = Bbox/10
num_walks = 4;


xvec1=np.zeros(0); yvec1=np.zeros(0); thetavec1 = np.zeros(0)
x1=1.0; y1 = 1.0; theta1 = (2*np.pi)*rand()
xvec2=np.zeros(0); yvec2=np.zeros(0); thetavec2 = np.zeros(0)
x2=1.0; y2 = -1.0; theta2 = (2*np.pi)*rand()
xvec3=np.zeros(0); yvec3=np.zeros(0); thetavec3 = np.zeros(0)
x3=-1.0; y3 = 1.0; theta3 = (2*np.pi)*rand()
xvec4=np.zeros(0); yvec4=np.zeros(0); thetavec4 = np.zeros(0)
x4=-1.0; y4 = -1.0; theta4 = (2*np.pi)*rand()
num_steps = 10000;

for i in range(num_steps):
    dx1 = np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dy1= np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dtheta1 = np.sqrt(2*Drot*dt)*(2*np.pi)*(rand(1)-0.5);
    x1 += vel*dt*np.cos(theta1) + dx1
    y1 += vel*dt*np.sin(theta1) + dy1
            
    if x1 > Bbox/2:
        x1 -= 2*(x1 - Bbox/2)
    elif x1 < -Bbox/2:
        x1 -= 2*(x1 + Bbox/2)
    elif x1 < Bbox/2 and x1 > -Bbox/2:
        x1 += 0
            
    if y1 > Bbox/2:
        y1 -= 2*(y1 - Bbox/2)
    elif y1 < -Bbox/2:
        y1 -= 2*(y1 + Bbox/2)
    elif y1 < Bbox/2 and y1 > -Bbox/2:
        y1 += 0
    xvec1 = np.append(xvec1,x1); yvec1 = np.append(yvec1,y1) 
    theta1 += dtheta1
    thetavec1 = np.append(thetavec1, theta1)

for i in range(num_steps):
    dx2 = np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dy2= np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dtheta2 = np.sqrt(2*Drot*dt)*(2*np.pi)*(rand(1)-0.5);
    x2 += vel*dt*np.cos(theta2) + dx2
    y2 += vel*dt*np.sin(theta2) + dy2
            
    if x2 > Bbox/2:
        x2 -= 2*(x2 - Bbox/2)
    elif x2 < -Bbox/2:
        x2 -= 2*(x2 + Bbox/2)
    elif x2 < Bbox/2 and x2 > -Bbox/2:
        x2 += 0
            
    if y2 > Bbox/2:
        y2 -= 2*(y2 - Bbox/2)
    elif y2 < -Bbox/2:
        y2 -= 2*(y2 + Bbox/2)
    elif y2 < Bbox/2 and y2 > -Bbox/2:
        y2 += 0
    xvec2 = np.append(xvec2,x2); yvec2 = np.append(yvec2,y2) 
    theta2 += dtheta2
    thetavec2 = np.append(thetavec2, theta2)

for i in range(num_steps):
    dx3 = np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dy3= np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dtheta3 = np.sqrt(2*Drot*dt)*(2*np.pi)*(rand(1)-0.5);
    x3 += vel*dt*np.cos(theta3) + dx3
    y3 += vel*dt*np.sin(theta3) + dy3
            
    if x3 > Bbox/2:
        x3 -= 2*(x3 - Bbox/2)
    elif x3 < -Bbox/2:
        x3 -= 2*(x3 + Bbox/2)
    elif x3 < Bbox/2 and x3 > -Bbox/2:
        x3 += 0
            
    if y3 > Bbox/2:
        y3 -= 2*(y3 - Bbox/2)
    elif y3 < -Bbox/2:
        y3 -= 2*(y3 + Bbox/2)
    elif y3 < Bbox/2 and y3 > -Bbox/2:
        y3 += 0
    xvec3 = np.append(xvec3,x3); yvec3 = np.append(yvec3,y3) 
    theta3 += dtheta3
    thetavec3 = np.append(thetavec3, theta3)
for i in range(num_steps):
    dx4 = np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dy4= np.sqrt(2*Dtrans*dt)*2*(rand(1)-0.5); 
    dtheta4 = np.sqrt(2*Drot*dt)*(2*np.pi)*(rand(1)-0.5);
    x4 += vel*dt*np.cos(theta4) + dx4
    y4 += vel*dt*np.sin(theta4) + dy4
            
    if x4 > Bbox/2:
        x4 -= 2*(x4 - Bbox/2)
    elif x4 < -Bbox/2:
        x4 -= 2*(x4 + Bbox/2)
    elif x4 < Bbox/2 and x4 > -Bbox/2:
        x4 += 0
            
    if y4 > Bbox/2:
        y4 -= 2*(y4 - Bbox/2)
    elif y4 < -Bbox/2:
        y4 -= 2*(y4 + Bbox/2)
    elif y4 < Bbox/2 and y4 > -Bbox/2:
        y4 += 0
    xvec4 = np.append(xvec4,x4); yvec4 = np.append(yvec4,y4) 
    theta4 += dtheta4
    thetavec4 = np.append(thetavec4, theta4)   

plt.plot(xvec1,yvec1, 'r-')
plt.plot(xvec2,yvec2, 'g-')
plt.plot(xvec3,yvec3, 'b-')
plt.plot(xvec4,yvec4, 'y-')

plt.xlim(-Bbox/2 - Bframe, Bbox/2 + Bframe)
plt.ylim(-Bbox/2 - Bframe, Bbox/2 + Bframe)
```

```python
walker1x = np.asarray(xvec1)
walker1y = np.asarray(yvec1)
walker2x = np.asarray(xvec2)
walker2y = np.asarray(yvec2)
walker3x = np.asarray(xvec3)
walker3y = np.asarray(yvec3)
walker4x = np.asarray(xvec4)
walker4y = np.asarray(yvec4)

df = pd.DataFrame({"X trajectory 1" : walker1x, "y trajectory 1" : walker1y,
                   "X trajectory 2" : walker2x, "y trajectory 2" : walker2y,
                  "X trajectory 3" : walker3x, "y trajectory 3" : walker3y,
                  "X trajectory 4" : walker4x, "y trajectory 4" : walker4y,})
df.to_csv("allwalkerdata.csv", index=False)
```
